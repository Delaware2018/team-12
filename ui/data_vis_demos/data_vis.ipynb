## Visualization Sketches

### 1. Basic Bubble Plot

For showing distribution - using matplotlib, numpy, and seaborn

Ref: <a href = "https://python-graph-gallery.com/341-python-gapminder-animation/">python graph gallery</a>

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
 
# create data
x = np.random.rand(15)
y = x+np.random.rand(15)
z = x+np.random.rand(15)
z=z*z
```

```python
# Generate and show bubble plot

#Change color with c and alpha. I map the color to the X axis value.
plt.scatter(x, y, s=z*2000, c=x, cmap="Blues", alpha=0.4, edgecolors="grey", linewidth=2)
 
# Add titles (main and on axis)
plt.xlabel("the X axis")
plt.ylabel("the Y axis")
plt.title("A colored bubble plot")
 
plt.show()
```

### 2. Density Hexagon Map
Ref: <a href="http://bokeh.pydata.org/en/latest/docs/gallery/hexbin.html">Hexbin.py</a>


```python
# libraries
import numpy as np
from bokeh.io import output_file, show, output_notebook
from bokeh.models import HoverTool
from bokeh.plotting import figure

# Load dataset (for now we are using random data points as filler)
n = 1000 # n is the number of data points
x = 2 + 2*np.random.standard_normal(n)
y = 2 + 2*np.random.standard_normal(n)
```

```python
# display customization
p = figure(title="Hexbin for 500 points", match_aspect=True,
           tools="wheel_zoom,reset", background_fill_color='#440154')
p.grid.visible = False

r, bins = p.hexbin(x, y, size=0.5, hover_color="pink", hover_alpha=0.8)

p.circle(x, y, color="white", size=1)

p.add_tools(HoverTool(
    tooltips=[("count", "@c"), ("(q,r)", "(@q, @r)")],
    mode="mouse", point_policy="follow_mouse", renderers=[r]
))
```

```python
# output_file("hexbin.html")
output_notebook()
show(p)
```
